```python
#1.Find the error in the following Python code:

 for i in range(5)
     print(i)
     
# ANSWER : SyntaxError - " : " missed after function
for i in range(5):
    print(i)
```

```python
#2.What will be the output of this code?

a = 10
b = "5"
print(a + b)

# ANSWER : TypeError - 'integer' & "string' couldn't be added '+'
a = 10
b = 5
print(a + b)
```

```python
#3.What is the issue in this Python snippet?

 def add_numbers(a, b)
    return a + b

# ANSWER : SyntaxError - indentation space before "def" and " : " missed after function
def add_numbers(a, b):
    return a + b

add_numbers(4,6)
```

```python
#4.Identify the bug in the following Python code:

numbers = [1, 2, 3, 4]
print(numbers[4])

# ANSWER : SyntaxError for indentation and IndexError labeled index from 0 to 3 for this list and it is out of range
numbers = [1, 2, 3, 4]
print(numbers[3])
```

```python
#7.Find the error in the following Python code:

name = "Alice"
age = 25
print("Name: " + name + ", Age: " + age)

# ANSWER : TypeError - missed comma after strings
name = "Alice"
age = 25
print("Name: " , name + ", Age: " , age)
```

```python
#8.Find the error

for i in range(1, 10, -1):
    print(i)

# ANSWER : IndexError - STEP must be >0 and not -1, list index out of range
for i in range(1, 10, 1):
    print(i)
```

```python
#9. Find the error in the following Python code:

x = 10
if x = 5:
    print("x is 5")

# ANSWER : SyntaxError - missed " == " after variable 'x'
x = 10
if x == 5:
    print("x is 5")
```
